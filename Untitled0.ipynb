```python
import numpy as np
import pandas as pd

# --- STEP 1: GENERATE DATA (The "Table") ---

# 1. Create 4000 "Boring" Stars
# 'g_r' is the Green-Red color. Stars are usually between -0.5 and 2.0.
stars_gr = np.random.uniform(-0.5, 2.0, 4000)
# Stars follow a pattern: Redder stars have more UV (u-g) in a predictable way.
stars_ug = (1.2 * stars_gr) + 0.8 + np.random.normal(0, 0.2, 4000)

# 2. Create 1000 "Monster" Quasars
# Quasars don't follow the rules. They are randomly scattered but generally have "UV Excess" (low u-g).
qso_gr = np.random.normal(0.3, 0.3, 1000)
qso_ug = np.random.normal(0.2, 0.4, 1000) # <--- This is the secret "Black Hole" signal

# 3. Combine them into one big "TOPCAT Table" (DataFrame)
# We label them so we can check our answers later.
df_stars = pd.DataFrame({'g_r': stars_gr, 'u_g': stars_ug, 'type': 'STAR'})
df_qso   = pd.DataFrame({'g_r': qso_gr,   'u_g': qso_ug,   'type': 'QUASAR'})

# Stack them on top of each other
data = pd.concat([df_stars, df_qso])

# Shuffle the deck so they are mixed up (like real data)
data = data.sample(frac=1).reset_index(drop=True)

print(f"Data Table Created! We have {len(data)} objects.")
print("Here are the first 5 rows (Notice we have mixed Stars and Quasars):")
print(data.head())
```

```python
# --- STEP 2: ORGANIZE THE DATA ---

# In TOPCAT, you would create "Subsets".
# In Python, we create "Variables".

# Let's verify what we have.
# This counts how many of each type are in our table.
counts = data['type'].value_counts()

print("Census of our Universe:")
print(counts)

# Separation logic:
# We want to put the Stars in one variable and Quasars in another
# so we can paint them different colors.
my_stars = data[data['type'] == 'STAR']
my_quasars = data[data['type'] == 'QUASAR']

print(f"\nWe are ready to plot {len(my_stars)} Stars and {len(my_quasars)} Black Holes.")
```

```python
import matplotlib.pyplot as plt

# --- STEP 3: THE PLOT ---

plt.figure(figsize=(10, 8))

# 1. Plot the "Boring" Stars first
# We use small black dots (alpha=0.1 makes them transparent so we see density)
plt.scatter(my_stars['g_r'], my_stars['u_g'],
            s=5, color='black', alpha=0.1, label='Normal Stars')

# 2. Plot the "Hidden" Quasars
# We use Red dots to make them pop out.
plt.scatter(my_quasars['g_r'], my_quasars['u_g'],
            s=10, color='red', alpha=0.6, label='Quasars (Black Holes)')

# 3. Decorate the Graph (Make it look professional)
plt.title("The 'Color-Color' Discovery Plot", fontsize=16)
plt.xlabel("Color (Green - Red)", fontsize=14)
plt.ylabel("Color (UV - Green)", fontsize=14)

# Set the limits (The "Zoom")
plt.xlim(-0.5, 2.5)
plt.ylim(-1.0, 3.5)

plt.legend()
plt.grid(True, alpha=0.3)

# 4. Show the result
plt.show()
```
